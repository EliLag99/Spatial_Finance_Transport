### **PreProcessing**

#### **Remove negative values**

```python
%cd ~/../content/yolov5_obb/dataset/VehSat

path = "/content/yolov5_obb/dataset/VehSat/labelTxt"
files = [f for f in listdir(path) if isfile(join(path, f))]

for file in files:

  # read file
  with open(path+"/"+file) as topo_file:
    lines = topo_file.readlines()

  # Write file
  with open(path+"/"+file, 'w') as fp:
    # iterate each line
    for number, line in enumerate(lines):
        # delete line 5 and 8. or pass any Nth line you want to remove
        # note list index starts from 0
        if '-' not in line:
            fp.write(line)
```
